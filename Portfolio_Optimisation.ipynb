```python
import pandas
import numpy
import seaborn
import scipy
import yfinance
```

```python
def FetchDataYahoo(Tickers,Columns=None,StartDate=None,StopDate=None,Timespan='max',Interval='1d'):
    """uses yfinance to get data for the array of tickers using yfinance download method.
    tickers : str, list
        List of tickers to download eg.['T','PG']
    Columns = list
        Array of columns needed. The columns are ['Adj Close','Close','High','Low','Open','Volume']
    StartDate = Download start date string 'YYYY-MM-DD' or _datetime.
    StopDate = Download end date string 'YYYY-MM-DD' or _datetime.
    period : str
        Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
        Either Use period parameter or use start and end
    interval : str
        Valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
        Intraday data cannot extend last 60 days
    
    """
    StockData= yfinance.download(Tickers,start=StartDate,end=StopDate,period=Timespan,interval=Interval)
    if Columns!=None:
        SelectedData = StockData.loc[:,Columns]
    else:
        SelectedData = StockData
    
    return(SelectedData)
```

```python
TickerData=FetchDataYahoo(['T','PG','K'],Columns=['Adj Close'],Timespan='1y')
```

```python
def CalcGain(Data,Tickers):
    
    for i in Tickers:
        temp = Data.loc[:,('Adj Close',i)]
        Data.loc[:,('Return',i)] = (temp-temp.shift(1))/temp[0]
    
    return(Data)
```

```python
CalcGain(TickerData,['T','PG','K'])
```
